## Question:2

###  How to stack two arrays vertically?

#### Desired Output::

```python
import numpy as np
```

```python
arr1 = np.arange(10).reshape(2,5)
arr2 = np.ones(10 , dtype = int).reshape(2,5)
print(arr1,arr2)
np.vstack((arr1,arr2))
```

## Question:3

### How to stack two arrays horizontally?

#### Desired Output::

```python
import numpy as np
```

```python
arr1 =  np.arange(10).reshape(2,5)
arr2 = np.ones(10,dtype= int).reshape(2,5)
print((arr1,arr2))
np.hstack((arr1,arr2))
```
